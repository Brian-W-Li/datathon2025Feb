## Cleaning

### PLAYER_ATTRIBUTES.csv Cleaning

```python
display(df_player_attributes.sort_values(by='aggression'))
```

```python
unique_nan = df_player_attributes.loc[df_player_attributes['aggression'].isna(), 'player_api_id'].unique()
completely_nan = np.array([])
for id in unique_nan:
    sum = 0
    for row in range(len(df_player_attributes)):
        if df_player_attributes.loc[row, 'player_api_id'] == id and not np.isnan(df_player_attributes.loc[row, 'aggression']):
            sum+=1
            break
    if sum == 0:
        np.insert(completely_nan, id)
print(completely_nan)
            
```

```python
df_player_attributes = df_player_attributes.dropna(subset=[df_player_attributes.columns[4]])
df_player_attributes = df_player_attributes.drop(columns=['player_fifa_api_id'])
display(df_player_attributes)
```

```python
df_player_attributes.to_csv('FINAL_PLAYER_ATTRIBUTES.csv', index=False)
```

### MATCH.csv Cleaning

Delete blank columns

```python
df_match = df_match.drop(columns=[col for col in df_match.columns if "Unnamed" in col])
display(df_match)
```

Delete rows that have more than 5 blank players for `home_team` or `away_team`

```python
df_match = df_match[df_match.iloc[:, 11:22].isna().sum(axis=1) <= 5]
df_match = df_match[df_match.iloc[:, 22:33].isna().sum(axis=1) <= 5]

display(df_match)
        
```

```python
df_match.to_csv('CLEANED1_european_soccer_dataset-MATCH.csv', index=False)
```

```python
for i in range(len(df_match)):
    df_match.iloc[i, 5] = df_match.iloc[i, 5].split(' ')[0]
for i in range(len(df_match)):
    date = df_match.iloc[i, 5].split('/')
    df_match.iloc[i, 5] = date[0] + '-' + date[1] + '-' + date[2]
display(df_match)
```
