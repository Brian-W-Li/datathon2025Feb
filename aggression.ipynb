## MATCH.CSV Cleaning

Delete blank columns

```python
df_match = df_match.drop(columns=[col for col in df_match.columns if "Unnamed" in col])
display(df_match)
```

Delete rows that have more than 5 blank players for `home_team` or `away_team`

```python
df_match = df_match[df_match.iloc[:, 11:22].isna().sum(axis=1) <= 5]
df_match = df_match[df_match.iloc[:, 22:33].isna().sum(axis=1) <= 5]

display(df_match)
        
```

```python
df_player_attributes = df_player_attributes.dropna(subset=[df_player_attributes.columns[2],df_player_attributes.columns[4]])
display(df_player_attributes)
```

```python
df_match.to_csv('CLEANED1_european_soccer_dataset-MATCH.csv', index=False)
```

```python
for i in range(len(df_match)):
    df_match.iloc[i, 5] = df_match.iloc[i, 5].split(' ')[0]
for i in range(len(df_match)):
    date = df_match.iloc[i, 5].split('/')
    df_match.iloc[i, 5] = date[0] + '-' + date[1] + '-' + date[2]
display(df_match)
```

```python
df_match.iloc[:,6:]=df_match.iloc[:,6:].dropna().astype('Int64')
display(df_match)
```
